# 分割 訓練資料 & 測試資料

```python
# 從 report.csv 分離出 預測資料 & 訓練資料

submission_ID = submission_set['ID'].values

test_report_set = report_set[ report_set["ID"].isin(submission_ID) ]
train_report_set = report_set[ ~report_set["ID"].isin(submission_ID) ]


print("submission_ID：{{ len = {submit_length} , array = {submit_ID} }} \n" .format(submit_ID = submission_ID, submit_length = len(submission_ID)))

"Original report_set："
report_set.head()

"test_report_set："
test_report_set.head()

"train_report_set："
train_report_set.head()


test_report_set.drop('ID', axis = 1 , inplace = True)
train_report_set.drop('ID', axis = 1 , inplace = True)
```

```python
# 指定目標變數、解釋變數
x_train = train_report_set.drop('乳量', axis = 1)
y_train = train_report_set['乳量']

x_test = test_report_set.drop('乳量', axis = 1)

"x_train"
x_train

"y_train"
y_train

"x_test"
x_test
```

# 線性回歸

```python
# 為了多元線性迴歸模型建構的導入
from sklearn.linear_model import LinearRegression

# 為了資料分割(訓練資料與測試資料)的匯入
from sklearn.model_selection import train_test_split

# 多元線性迴歸的初始化學習
model = LinearRegression()
model.fit(x_train, y_train)

# 顯示決定係數
print ('決定係數(train)：{:.3f}' .format(model.score(x_train, y_train)) )

result = model.predict(x_test)
submission_set['預測乳量'] = result.tolist()
submission_set.rename(columns = {'預測乳量':'1'} , inplace = True)
submission_set

submission_set.to_csv("Submission.csv", index = 0)
```
